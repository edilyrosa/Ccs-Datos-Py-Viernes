**1. CREAMOS Y MOSTRAMOS EL DATAFRAME**

```
import pandas as pd
import numpy as np

data = {
    'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05','2023-01-01', '2023-01-06', '2023-01-07', '2023-01-08',
             '2023-01-09','2023-01-03', '2023-01-03'],
    'Location': ['CityA', 'CityB', 'CityA', 'CityC', 'CityB', 'CityA', 'CityD', 'CityA', 'CityE', 'CityC', 'CityA', 'faltante'],
    'TemperatureChange': [1.2, 0.5, None, 2.1, 1.5, 1.2, 0.8, 1.9, None, 2.5, None, None],
    'Humidity': [60, 65, 70, None, 75, 60, 80, 85, 90, None, 70, None],
    'Notes': ['Clear', 'Cloudy', 'Rainy', 'Sunny', 'Clear', 'Clear', 'Cloudy', 'Rainy', 'Sunny', 'Clear', 'Rainy', None],
    'MixedColumn': [10, 'twenty', 30, 40.5, 'fifty', 60, 70, 'eighty', 90, 100, 30, None]
}

#1. CREAMOS EL DATAFRAME
df = pd.DataFrame(data)
#2. VISUIALIZACION DEL DF
display(df)

```

image.png

```
#Visulizacion de los regustros duplicados.
duplicate_rows = df[df.duplicated()]
print('Mostrando las filas o registrso duplicados')
display(duplicate_rows)

#Eliminacion de los regustros duplicados.
df_cleaned = df.drop_duplicates(keep='first') #Elimino todos los duplicados y consev la primera aparicion

print('Dataframe sin duplicados')
display(df_cleaned)
```

image.png

```
import seaborn as sns
import matplotlib.pyplot as plt

#4. Visualizacion de los valores faltantes.
plt.figure(figsize=(10, 6))
sns.heatmap(df_cleaned.isnull(), cbar=False, cmap='viridis')
plt.title('Mapa de calor de valores faltantes')
plt.show()

#SUMANDO VALORES FALTATES POR COLUMNA
valores_faltantes = df_cleaned.isnull().sum()
display(valores_faltantes)
```

image.png

```
#5 filas o registros con demasiados datos faltantes.
non_missing_counts = df_cleaned.count(axis=1)

#si le faltan mas de la mitad la eliminamos.
limite = df_cleaned.shape[1]/2

df_filtered = df_cleaned[non_missing_counts > limite].copy()
print('Mostrando nuevo df filtrado sin registros muy flatantes de data')
display(df_filtered)

```

image.png

```
#6. RELLENAR LOS DATOS FALTANTES.
#convertir a tipo de dato "datetime" la columna "date"
df_filtered['Date'] = pd.to_datetime(df_filtered['Date'])
#ordenando el df por fecha
df_filtered = df_filtered.sort_values(by='Date')

#Rellenar los valores faltantes (NAN) con las columnas que los tienen: TemperatureChange	Humidity
df_filtered[['TemperatureChange',	'Humidity']] = df_filtered[['TemperatureChange',	'Humidity']].interpolate(method='linear')
print('Dataframe con los datos faltantes rellenados en las col TemperatureChange	Humidity	con interpolacion lineal')
display(df_filtered)
```

image.png

```
#7 REGULARIZAR EL TDD DE LA COL "MixedColumn"
df_filtered['MixedColumn'] = pd.to_numeric(df_filtered['MixedColumn'], errors='coerce')

#rellenamos los Nan con la media
df_filtered['MixedColumn'].fillna(df_filtered['MixedColumn'].mean(), inplace=True)
print('Los TDD han sido regularizados')
display(df_filtered.dtypes)

#Mostrsmos los primeros registros.
display(df_filtered.head())
```

image.png

```
# 8 . analisis estadistico descriptivo de las columnas numericas.
print('analisis estadistico descriptivo de las columnas numericas.')
display(df_filtered.describe())

#mostrar y contar los valores unicos en las columnas categoricas: Location y Notes
for col in ["Location",  "Notes"]:
  print(f'Valores {col}: ')
  display(df_filtered[col].value_counts())
```

image.png

```
#10 calcular y visualizar los outlers.
#selec las col a aanalizar.
colums_to_analyze = ["TemperatureChange", "Humidity", "MixedColumn"]
#calcular q1, q3 y IQR
Q1 = df_filtered[colums_to_analyze].quantile(0.25)
Q3 = df_filtered[colums_to_analyze].quantile(0.75)
IQR = Q3 - Q1

# determianndo los outlers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print('OUTLER MAS BAJO', lower_bound)
print('OUTLER MAS ALTO', upper_bound)

#Identificar los outlears
outlears = {}
for col in colums_to_analyze:
  col_outlers  = df_filtered[(df_filtered[col] < lower_bound[col]) | (df_filtered[col] > upper_bound[col])]
  outlears[col] = col_outlers

print('OUTLER POTENCIALES')
for col, col_outlers in outlears.items():
  print(f' {col}: {len(col_outlers)}')

#Boxplot donde podemos visualizar los aoutlers
plt.figure(figsize=(15, 5))
for i, col in enumerate(colums_to_analyze):
  plt.subplot(1, 3, i + 1)
  sns.boxplot(x=df_filtered[col])
  plt.title(f'Boxplot {col}')
  plt.tight_layout()
  plt.show()


```
