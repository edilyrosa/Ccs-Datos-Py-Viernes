**I. 游댝DataFrame con tempetaruta Sinusoidal**

```python
import pandas as pd
import numpy as np
import random

# N칰mero de horas en un d칤a
num_horas = 24

# Datos para las columnas
ciudad = [f'Ciudad-{num}' for num in range(num_horas)]
hora = list(range(num_horas))

temperatura_list = []
humedad_list = []

# Generar patrones base para temperatura y humedad que cambian gradualmente
# Temperatura: sigue un patr칩n sinusoidal, con pico en la tarde
# Patr칩n sinusoidal: Significa que la temperatura aumenta gradualmente, alcanza un punto m치ximo,
# y luego disminuye gradualmente hasta un punto m칤nimo, antes de volver a empezar el ciclo.
# No hay cambios bruscos, sino una oscilaci칩n regular.
# Humedad: sigue un patr칩n inversamente sinusoidal al de la temperatura
for h in range(num_horas):
    # Patr칩n de temperatura: empieza baja, sube al mediod칤a/tarde, baja en la noche
    # 20: Temperatura media/base (promedio del d칤a).
    # 10: Amplitud (oscila 췀10춿C alrededor de 20춿C, rango 10-30춿C).
    # np.sin(): Funci칩n seno que crea la onda suave (sube/baja gradualmente).
      # El `h - 8` desplaza el pico para que est칠 alrededor de la hora 14 (8+6).
      # np.sin(...): seno que produce una onda que oscila entre -1 y 1
      # np.pi / 12 escala el eje de las horas (h) para que la onda sinusoidal complete un ciclo
      # completo (de 0 a 2 radianes, que es la duraci칩n de un ciclo completo del seno) exactamente en 24
      # horas (2 * 12 = 24). Es decir, asegura que el patr칩n se repita diariamente.
    temp_base_value = 20 + 10 * np.sin((h - 8) * np.pi / 12)

    # Patr칩n de humedad: inversamente relacionado con la temperatura
    hum_base_value = 60 - 20 * np.sin((h - 8) * np.pi / 12)

    # A침adir una peque침a fluctuaci칩n aleatoria (ruido)
    current_temp = temp_base_value + random.uniform(-2, 2) # r.uniform(-2, 2) genera float entre -2 y 2
    current_hum = hum_base_value + random.uniform(-5, 5)

    # Asegurarse de que los valores est칠n dentro de rangos razonables
    current_temp = max(10, min(35, current_temp))
    current_hum = max(30, min(95, current_hum))

    #llenamos el array con la temperatura
    # 游댠SIN NAN: temperatura_list.append(round(current_temp, 1))

    # 游댠CON NAN: Introducir valores None en horas espec칤ficas para temperatura
    if h in [3, 10, 16]:
        temperatura_list.append(None)
    else:
        temperatura_list.append(round(current_temp, 1))

    #llenamos el array con la humedad
    humedad_list.append(round(current_hum, 1))

# Crear el DataFrame
df_clima = pd.DataFrame({
    'ciudad': ciudad,
    'hora': hora,
    'temperatura': temperatura_list,
    'humedad': humedad_list
})

display(df_clima)

# Para verificar los valores nulos que se crearon
display('Tipo de dato por columna:')
display(df_clima.dtypes)
```

```python
#*********************************** 游댝 IMPUTACION CON VALORES CENTRALES.
# #游댠Rellenamos NaN de "temperatura" y "humedad" con la media de la columna
# df_clima['temperatura'] = df_clima['temperatura'].fillna(df_clima['temperatura'].mean())

#*********************************** 游댝 IMPUTACION CON INTERPOLACION.
# #游댠Metodo de Interpolacion lineal
# df_clima['temperatura'] = df_clima['temperatura'].interpolate(method='linear', limit_direction='both')

# #游댠Metodo de Interpolacion Spline
df_clima['temperatura'] = df_clima['temperatura'].interpolate(method='spline', order=3, limit_direction='both').round(2)
```

游늵**2. Visualizacion de la data, donde se aprecia la onda sinusoidal**

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

# Plot temperature
plt.plot(df_clima['hora'], df_clima['temperatura'], label='Temperatura (춿C)', color='red', marker='o', linestyle='-')

# Plot humidity
plt.plot(df_clima['hora'], df_clima['humedad'], label='Humedad (%)', color='blue', marker='x', linestyle='--')

plt.title('Patr칩n de Temperatura y Humedad a lo largo del D칤a')
plt.xlabel('Hora del D칤a')
plt.ylabel('Valor')
plt.xticks(df_clima['hora'])
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```
