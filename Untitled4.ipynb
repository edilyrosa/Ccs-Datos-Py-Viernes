```python
import pandas as pd
import numpy as np
import random

# N√∫mero de horas en un d√≠a
num_horas = 24

# Datos para las columnas
ciudad = [f'Ciudad-{num}' for num in range(num_horas)]
hora = list(range(num_horas))

temperatura_list = []
humedad_list = []

# Generar patrones base para temperatura y humedad que cambian gradualmente
# Temperatura: sigue un patr√≥n sinusoidal, con pico en la tarde
# Patr√≥n sinusoidal: Significa que la temperatura aumenta gradualmente, alcanza un punto m√°ximo,
# y luego disminuye gradualmente hasta un punto m√≠nimo, antes de volver a empezar el ciclo.
# No hay cambios bruscos, sino una oscilaci√≥n regular.
# Humedad: sigue un patr√≥n inversamente sinusoidal al de la temperatura
for h in range(num_horas):
    # Patr√≥n de temperatura: empieza baja, sube al mediod√≠a/tarde, baja en la noche
    # 20: Temperatura media/base (promedio del d√≠a).
    # 10: Amplitud (oscila ¬±10¬∞C alrededor de 20¬∞C, rango 10-30¬∞C).
    # np.sin(): Funci√≥n seno que crea la onda suave (sube/baja gradualmente).
      # El `h - 8` desplaza el pico para que est√© alrededor de la hora 14 (8+6).
      # np.sin(...): seno que produce una onda que oscila entre -1 y 1
      # np.pi / 12 escala el eje de las horas (h) para que la onda sinusoidal complete un ciclo
      # completo (de 0 a 2œÄ radianes, que es la duraci√≥n de un ciclo completo del seno) exactamente en 24
      # horas (2 * 12 = 24). Es decir, asegura que el patr√≥n se repita diariamente.
    temp_base_value = 20 + 10 * np.sin((h - 8) * np.pi / 12)

    # Patr√≥n de humedad: inversamente relacionado con la temperatura
    hum_base_value = 60 - 20 * np.sin((h - 8) * np.pi / 12)

    # A√±adir una peque√±a fluctuaci√≥n aleatoria (ruido)
    current_temp = temp_base_value + random.uniform(-2, 2) # r.uniform(-2, 2) genera float entre -2 y 2
    current_hum = hum_base_value + random.uniform(-5, 5)

    # Asegurarse de que los valores est√©n dentro de rangos razonables
    current_temp = max(10, min(35, current_temp))
    current_hum = max(30, min(95, current_hum))

    # Introducir valores None en horas espec√≠ficas para temperatura
    if h in [3, 10, 16]:
        temperatura_list.append(None)
    else:
        temperatura_list.append(round(current_temp, 1))

    # Mantener la introducci√≥n de valores None aleatoria para humedad
    humedad_list.append(None if random.random() < 0.15 else round(current_hum, 1))

# Crear el DataFrame
df_clima = pd.DataFrame({
    'ciudad': ciudad,
    'hora': hora,
    'temperatura': temperatura_list,
    'humedad': humedad_list
})

display(df_clima)

# Para verificar los valores nulos que se crearon
display('Valores nulos por columna:')
display(df_clima.isnull().sum())
```

```python
import pandas as pd
import numpy as np
import random

# N√∫mero de horas en un d√≠a
num_horas = 24

# Datos para las columnas
sensor = ['sensor-00' if num in [3, 10, 16] else f'sensor-{num}' for num in range(num_horas)]

# (None if random.random() < 0.15 else round(current_hum, 1))
hora = list(range(num_horas))

temperatura_list = []
humedad_list = []

# Generar patrones base para temperatura y humedad que cambian gradualmente
# Temperatura: sigue un patr√≥n sinusoidal, con pico en la tarde
# Humedad: sigue un patr√≥n inversamente sinusoidal al de la temperatura
for h in range(num_horas):
    # üî•Patr√≥n de temperatura: empieza baja, sube al mediod√≠a/tarde, baja en la noche
    # Amplitud (oscila ¬±10¬∞C alrededor de 20¬∞C, rango 10-30¬∞C).
    # np.sin(): Funci√≥n seno que crea la onda suave (sube/baja gradualmente).
    temp_base_value = 20 + 10 * np.sin((h - 8) * np.pi / 12)

    # üî•Patr√≥n de humedad: inversamente relacionado con la temperatura
    hum_base_value = 60 - 20 * np.sin((h - 8) * np.pi / 12)

    # ‚öñÔ∏è A√±adir una peque√±a fluctuaci√≥n aleatoria (ruido)
     # r.uniform(-2, 2) genera float entre -2 y 2
    current_temp = temp_base_value + random.uniform(-2, 2)
    current_hum = hum_base_value + random.uniform(-5, 5)

    # üî•Asegurarse de que los valores est√©n dentro de rangos razonables {entre 10 y 35}
    # min() devuelve el valor m√°s peque√±o entre sus argumentos ‚Üí {35, current_tem}
    # max() devuelve el valor m√°s grande entre ‚Üí {10, el mas grande permitido=35}
    current_temp = max(10, min(35, current_temp))
    current_hum = max(30, min(95, current_hum))

    # ‚ùå Introducir valores None en horas espec√≠ficas para temperatura
    if h in [3, 10, 16]:
        temperatura_list.append(None)
    else:
      # Agrega el valor redondeado
        temperatura_list.append(round(current_temp, 1))

    # ‚ùå Introducir valores None aleatoria para humedad
    # Si el n√∫mero aleatorio generado es menor que 0.15 ‚Üí 0.01,se eval√∫a:
    # TRUE: agrega un None None a humedad_list.
    # FALSE: agrega el valor redondeado de la humedad actual a humedad_list.
    humedad_list.append(None if random.random() < 0.15 else round(current_hum, 1))

# üî•Crear el DataFrame
df_clima = pd.DataFrame({
    'sensor': sensor,
    'hora': hora,
    'temperatura': temperatura_list,
    'humedad': humedad_list
})

display(df_clima)

print('\n üìãSerie de TDD de los Valores de columnas numericas del DF')
display(df_clima[['temperatura', 'humedad' ]].dtypes) # Es un Attrb

# ‚ùåVerificamos los valores nulos que se crearon por columna del DF
print('\n ‚ùå Serie de Valores nulos por columna:')
display(df_clima.isnull().sum())
```

Claro, te explicar√© por qu√© se us√≥ interpolate(method='linear') y en qu√© se diferencia de nearest.

Interpolaci√≥n Lineal (method='linear')
La interpolaci√≥n lineal estima los valores faltantes asumiendo que hay una relaci√≥n lineal entre los puntos de datos conocidos. En otras palabras, traza una l√≠nea recta entre el valor conocido anterior y el valor conocido siguiente para inferir el valor intermedio.

¬øC√≥mo funciona? Si tienes un punto de datos (x1, y1) y otro (x2, y2), y necesitas estimar un valor y en un punto x entre x1 y x2, la interpolaci√≥n lineal calcula y como una media ponderada de y1 y y2 bas√°ndose en la proximidad de x a x1 y x2.

¬øPor qu√© se us√≥ aqu√≠? Se eligi√≥ linear porque los datos de temperatura y humedad se generaron siguiendo un patr√≥n sinusoidal. Un patr√≥n sinusoidal implica cambios suaves y graduales a lo largo del tiempo. La interpolaci√≥n lineal respeta esta gradualidad, rellenando los valores faltantes de manera que la transici√≥n sea continua y coherente con la tendencia general de los datos.

Interpolaci√≥n por Vecino M√°s Cercano (method='nearest')
La interpolaci√≥n por vecino m√°s cercano (o nearest) estima los valores faltantes asign√°ndoles el valor del punto de datos conocido m√°s cercano (ya sea anterior o posterior).

¬øC√≥mo funciona? Si tienes un valor faltante, simplemente busca el valor v√°lido m√°s cercano en el conjunto de datos (por ejemplo, en la fila anterior o siguiente) y lo asigna al valor faltante.

¬øPor qu√© no se us√≥ aqu√≠? Aunque nearest es simple, no es ideal para datos que tienen una tendencia o un patr√≥n suave como el sinusoidal. Al asignar el valor de un √∫nico vecino, puede introducir saltos o escalones bruscos en los datos donde deber√≠an haber transiciones suaves. Esto podr√≠a distorsionar el patr√≥n sinusoidal y no reflejar con precisi√≥n la naturaleza real de las mediciones de temperatura y humedad.

Diferencias clave y cu√°ndo usar cada uno:
linear: Ideal para datos con tendencias suaves, como series temporales, o datos que se espera que cambien de forma continua. Mantiene la continuidad y la tendencia general.
nearest: M√°s adecuado cuando los datos son discretos o cuando los cambios bruscos son esperados o aceptables. Puede ser √∫til en escenarios donde el valor faltante se asume id√©ntico al punto de datos m√°s cercano sin ninguna suposici√≥n de tendencia.
En resumen, para los datos de clima con un comportamiento sinusoidal, la interpolaci√≥n lineal ofrece una estimaci√≥n m√°s realista y consistente con el modelo subyacente de los datos.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generar el componente sinusoidal ideal para 24 horas
num_horas = 24
horas = np.arange(num_horas)

# El argumento de la funci√≥n seno es (h - 8) * np.pi / 12
# Donde 'h' es la hora del d√≠a. Esto crea un ciclo completo en 24 horas
# y desplaza el pico para que coincida con el patr√≥n original.

# Componente sinusoidal para la temperatura: np.sin(argumento)
# (La f√≥rmula original era 20 + 10 * sin_componente_temperatura)
sin_component_temp = np.sin((horas - 8) * np.pi / 12)

# Componente sinusoidal para la humedad: -np.sin(argumento)
# (La f√≥rmula original era 60 - 20 * sin_componente_humedad,
#  lo que significa que el componente seno se invert√≠a y se multiplicaba por 20)
sin_component_hum = -np.sin((horas - 8) * np.pi / 12)

# Crear un DataFrame temporal para facilitar la graficaci√≥n
df_ideal_sin = pd.DataFrame({
    'hora': horas,
    'sin_componente_temperatura': sin_component_temp,
    'sin_componente_humedad': sin_component_hum
})

plt.figure(figsize=(12, 6))
sns.lineplot(x='hora', y='sin_componente_temperatura', data=df_ideal_sin, label='Componente Sinusoidal (Temperatura)')
sns.lineplot(x='hora', y='sin_componente_humedad', data=df_ideal_sin, label='Componente Sinusoidal (Humedad)')

plt.title('Onda Sinusoidal Ideal de Temperatura y Humedad (Picos -1 a 1)')
plt.xlabel('Hora del D√≠a')
plt.ylabel('Valor Sinusoidal')
plt.xticks(range(24)) # Mostrar todas las horas en el eje X
plt.yticks(np.arange(-1, 1.1, 0.5)) # Asegurar que el eje Y muestre -1, 0, 1
plt.axhline(0, color='gray', linestyle='--', linewidth=0.8) # L√≠nea en 0 para referencia
plt.axhline(1, color='red', linestyle=':', linewidth=0.8) # L√≠nea en 1 para referencia de pico
plt.axhline(-1, color='blue', linestyle=':', linewidth=0.8) # L√≠nea en -1 para referencia de valle
plt.grid(True)
plt.legend()
plt.show()
```
